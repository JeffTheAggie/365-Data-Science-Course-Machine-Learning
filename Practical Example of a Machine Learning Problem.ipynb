```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()

#Load the data
raw_data = pd.read_csv("1.04. Real-life example.csv")
raw_data.head()
```

```python
#Preprocessing the data
#5 Number summary
raw_data.describe(include = "all")
```

```python
#Determining the variable of interest (our model is supposed to inspect how each variable effects the price of a car)
data = raw_data.drop(["Model"], axis = 1)
data.describe(include = "all")
```

```python
#Dealing with missing values
data.isnull().sum()
```

```python
#Deleting columns with missing values
data_no_mv = data.dropna(axis = 0)
data_no_mv.describe(include = "all")
```

```python
#Exploring the Probability Distribution Functions
sns.distplot(data_no_mv["Price"])
```

```python
#Dealing with outliers
q = data_no_mv["Price"].quantile(0.99)
data_1 = data_no_mv[data_no_mv["Price"] < q]
data_1.describe(include = "all")
```

```python
sns.distplot(data_1["Price"])
```

```python
sns.distplot(data_no_mv["Mileage"])
```

```python
q = data_1["Mileage"].quantile(0.99)
data_2 = data_1[data_1["Mileage"] < q]
sns.distplot(data_2["Mileage"])
```

```python
sns.distplot(data_no_mv["EngineV"])
```

```python
data_3 = data_2[data_2["EngineV"] < 6.5]
sns.distplot(data_3["EngineV"])
```

```python
sns.distplot(data_no_mv["Year"])
```

```python
q = data_3["Year"].quantile(0.01)
data_4 = data_3[data_3["Year"] > q]
sns.distplot(data_4["Year"])
```

```python
#Finally cleaning the data
data_cleaned = data_4.reset_index(drop = True)
data_cleaned.describe(include = "all")
```

```python
#Checking the Ordinary Least Squares Assumptions
f,(ax1, ax2, ax3) = plt.subplots(1, 3, sharey = True, figsize = (15,3))
ax1.scatter(data_cleaned["Year"], data_cleaned["Price"])
ax1.set_title("Price and Year")
ax2.scatter(data_cleaned["EngineV"], data_cleaned["Price"])
ax2.set_title("Price and EngineV")
ax3.scatter(data_cleaned["Mileage"], data_cleaned["Price"])
ax3.set_title("Price and Mileage")
plt.show()
```

```python
sns.distplot(data_cleaned["Price"])
```

```python
#Relaxing the assumptions
log_price = np.log(data_cleaned["Price"])
data_cleaned["log_price"] = log_price
data_cleaned
```

```python
f,(ax1, ax2, ax3) = plt.subplots(1, 3, sharey = True, figsize = (15,3))
ax1.scatter(data_cleaned["Year"], data_cleaned["log_price"])
ax1.set_title("Log Price and Year")
ax2.scatter(data_cleaned["EngineV"], data_cleaned["log_price"])
ax2.set_title("Log Price and Engine Volume")
ax3.scatter(data_cleaned["Mileage"], data_cleaned["log_price"])
ax3.set_title("Log Price and Mileage")
plt.show()
```

```python
data_cleaned = data_cleaned.drop(["Price"], axis = 1)

#Multicollinearity
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[["Mileage", "Year", "EngineV"]]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(["Year"], axis = 1)

#Create dummy variables
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
data_with_dummies.head()
```

```python
#Rearrange the columns
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

```python
#Linear Regression model

#Declare the inputs and the targets
targets = data_preprocessed["log_price"]
inputs = data_preprocessed.drop(["log_price"], axis = 1)

#Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)

#Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 365)

#Create the Linear Regression Model
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#Plotting the targets against the predictions
y_hat = reg.predict(x_train)
plt.scatter(y_train, y_hat)
plt.xlabel("Targets (y_train)", size = 18)
plt.ylabel("Predictions (y_hat)", size = 18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
#Residual plot
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size = 18)
```

```python
#R-Squared
reg.score(x_train, y_train)
```

```python
#Finding the weights and bias
#Intercept
reg.intercept_
```

```python
#Coefficents
reg.coef_
```

```python
#Summary table
reg_summary = pd.DataFrame(inputs.columns.values, columns = ["Features"])
reg_summary["Weights"] = reg.coef_
reg_summary
```

```python
data_cleaned["Brand"].unique()
```

```python
#Testing the data
y_hat_test = reg.predict(x_test)
plt.scatter(y_test, y_hat_test, alpha = 0.2)
plt.xlabel("Targets (y_test)", size = 18)
plt.ylabel("Predictions (y_hat_test)", size = 18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ["Predictions"])
df_pf.head()
```

```python
df_pf["Target"] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop = True)
y_test.head()
```

```python
df_pf["Target"] = np.exp(y_test)
df_pf.head()
```

```python
df_pf["Residual"] = df_pf["Target"] - df_pf["Predictions"]
df_pf["Difference%"] = np.absolute(df_pf["Residual"]/df_pf["Target"]*100)
df_pf
```
